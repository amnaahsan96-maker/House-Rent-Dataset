# Visualization

**Single line Graph**

**Rent trend (index wise)**

```python
import matplotlib.pyplot as plt

plt.plot(hr['Rent'])
plt.xlabel('House Index')
plt.ylabel('Rent')
plt.title('Rent Trend')
plt.show()
```

**Multiple Line Graph**

**Rent vs Size**

```python
plt.plot(hr['Rent'],label='Rent')
plt.plot(hr['Size'],label='Size')
plt.xlabel('House Index')
plt.ylabel('Rent/Size')
plt.title('Rent vs Size')
plt.legend()
plt.show()
```

**Bar Graph**


```python
hr['Furnishing Status'].value_counts().plot(kind='bar')
plt.title('Furnishing Status Distribution')
plt.xlabel('Furnishing Status')
plt.ylabel('Number of Houses')
plt.show()
```

**Histrogram**

```python
plt.hist(hr['Rent'], bins=10, color='lightgreen', edgecolor='black')
plt.xlabel('Rent')
plt.ylabel('Frequency')
plt.title('Rent Distribution')
plt.show()
```

**PIE CHARTS**

```python
hr['Furnishing Status'].value_counts().plot(
    kind='pie', autopct='%1.1f%%', startangle=90, colors=['lightgreen', 'lightpink','yellow']
)
plt.title('Furnishing Status Distribution')
plt.ylabel('')
plt.show()
```

**Scatter Graph**

```python
plt.scatter(hr['Size'], hr['Rent'])
plt.xlabel('Size (sqft)')
plt.ylabel('Rent')
plt.title('Rent vs Size')
plt.show()
```
